#Data Preprocessing

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/content/drive/MyDrive/breast_cancer_survival - breast_cancer_survival_.csv.csv')
df.head()
```

```python
df['Gender'].value_counts()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.duplicated().sum()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df[['Surgery_day', 'Surgery_month', 'Surgery_year']] = df['Date_of_Surgery'].str.split('-', expand=True)
```

```python
df.head()
```

```python
df.drop('Date_of_Surgery', axis=1, inplace=True)
```

```python
df.head()
```

```python
df[['Last_visit_day','Last_visit_month','Last_visit_year']] = df['Date_of_Last_Visit'].str.split('-', expand=True)
```

```python
df.head()
```

```python
df.drop('Date_of_Last_Visit', axis=1, inplace=True)
```

```python
df.head()
```

```python
df['Surgery_type'].value_counts()
```

#Label Encoding

```python
months={
    'Jan':1,
    'Feb':2,
    'Mar':3,
    'Apr':4,
    'May':5,
    'Jun':6,
    'Jul':7,
    'Aug':8,
    'Sep':9,
    'Oct':10,
    'Nov':11,
    'Dec':12
}

df['Surgery_month'].unique()
```

```python
df['Surgery_month'] = df['Surgery_month'].map(months)
df['Last_visit_month'] = df['Last_visit_month'].map(months)
```

```python
df.head()
```

```python
df['Patient_Status']=df['Patient_Status'].map({'Alive':1,'Dead':0})
```

```python
df['Tumour_Stage'].unique()
df['Tumour_Stage']=df['Tumour_Stage'].map({'I':1,'II':2,'III':3})
```

```python
df.head()
```

```python
df.head()
```

```python
df['ER status'].unique()
```

```python
df.drop('ER status', axis=1, inplace=True)
```

```python
df['PR status'].unique()
```

```python
df.drop('PR status', axis=1, inplace=True)
```

```python
df['HER2 status'].unique()
```

```python
df['HER2 status']=df['HER2 status'].map({'Positive':1,'Negative':0})
```

```python
df.head()
```

```python
df['Histology'].unique()
```

```python
df['Histology']=df['Histology'].map({'Infiltrating Ductal Carcinoma':1,'Infiltrating Lobular Carcinoma':2,'Mucinous Carcinoma':3})
```

```python
df['Gender'].unique()
```

```python
df['Gender']=df['Gender'].map({'MALE':0,'FEMALE':1})
```

```python
df['Surgery_type'].unique()
```

```python
df['Surgery_type']=df['Surgery_type'].map({'Other':0,'Lumpectomy':1,'Modified Radical Mastectomy':2,'Simple Mastectomy':3})
```

```python
df.head()
```

```python
df['Gender'].unique()
```

#Correlation heatmap

```python
plt.figure(figsize=(15,15))
sns.heatmap(df.corr(), annot=True, cmap='cool')
```

#According to the correlation heatmap:
We see a positive correlation between surgery year and last visit year, this tells us that:

After the surgery, approximately 50% of the patients did not require medical attention.

However, more than half of the people still seek follow-up.

#Age distribution graph

```python
plt.figure(figsize=(10, 6))
sns.displot(x='Age', data=df)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

#The age distribution graph shows that:

People around the age of 45-65 are majority the victims of breast cancer.

#Patient Status count

```python
sns.countplot(x='Patient_Status', data=df)
```

#From this countplot, we see that:

Majority of the patients are alive as compared to the ones who are dead, so we can arrive at the conclusion that breast cancer is treatable and has a good survival rate if detected and treated early.

#Comparision of Tumour stage VS Patient status

```python
fig = plt.figure(figsize=(10,6))

sns.countplot(x='Tumour_Stage', hue='Patient_Status', data=df)
```

From this chart, we can draw the conclusion that, major patients are in the Tumour stage 2 category.

Also, on comparing stages 1 and 3, higher the stage, higher is the death ratio, so early detection of breast cancer is better and can be recovered.

Most of the patients discover their breast cancer in stage 2, in which nearly 1/4th of the people have the risk of being dead.

#Year wise surgery count

```python
plt.figure(figsize=(10,6))
sns.countplot(x='Surgery_year', data=df)
plt.title('Year wise surgeries')
plt.xlabel('Surgery year')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

We can see that most of the surgeries were held in the year 2018.

The number was to less than half in the year 2017, than 2018.

In the year 2019, it surged to nearly 120, which is more than 2017, but less than 2018.

#Year wise deaths

```python
sns.countplot(x='Surgery_year', hue='Patient_Status', data=df)
plt.title('Year wise deaths')
plt.xlabel('Surgery year')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

On seeing the above graph and this graph, there were around 150 surgeries done, still there were 30 people who died in the year 2018

The patients were less in the year 2017,

In the year 2019, higher number of deaths were recorded.
